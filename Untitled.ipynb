```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fe=np.loadtxt("Exp_1/Fe.Asc")
```

```python
plt.figure(figsize=(20,10))
plt.plot(Fe[:,0],Fe[:,1],'-x')
plt.xlabel("道址",size=20)
plt.ylabel("道计数",size=20)
plt.xlim([0,511])
plt.ylim([500,900])
plt.title(r"$\alpha-\rm{Fe}$",size=30)
plt.savefig("Exp_1/Fe.png",ppi=300)
```

```python
NaX=np.loadtxt("Exp_1/NAX.Asc")
plt.figure(figsize=(20,10))
plt.plot(NaX[:,0],NaX[:,1],'-x')
plt.xlabel("道址",size=20)
plt.ylabel("道计数",size=20)
print(plt.ylim())
plt.xlim([0,511])
plt.ylim([600,1060])
plt.title(r"$\rm{NaX}$",size=30)
plt.savefig("Exp_1/NAX.png",ppi=300)
```
